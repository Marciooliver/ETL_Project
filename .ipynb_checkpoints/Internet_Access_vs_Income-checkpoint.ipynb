```python
import pandas as pd 
import numpy as np
from sqlalchemy import create_engine
```

```python
internet_income = "resources/Internet Usage vs income per capita.csv"
internet_income_df = pd.read_csv(internet_income)
clean_internet_income_df = internet_income_df.drop([0,1,2,3,9,12,16,20,24,25,27,30,32,33,35,38,39,44,45])
clean_internet_income_df.head()
```

```python
internet_income_transformed = clean_internet_income_df [["Grandes Regiões, Unidades da Federação e Regiões Metropolitanas", "Rendimento médio mensal domiciliar per capita das pessoas de 10 anos ou mais de idade (R$)", "Unnamed: 2", "Unnamed: 3", "Rendimento mediano mensal domiciliar per capita das pessoas de 10 anos ou mais de idade (R$)","Unnamed: 5","Unnamed: 6"]]
internet_income_transformed = internet_income_transformed.rename(columns={"Grandes Regiões, Unidades da Federação e Regiões Metropolitanas": "Brazilian_States", "Rendimento médio mensal domiciliar per capita das pessoas de 10 anos ou mais de idade (R$)": "Income_Average_Per_Capita", "Unnamed: 2": "With_Access_Ave", "Unnamed: 3": "Without_Access_Ave", "Rendimento mediano mensal domiciliar per capita das pessoas de 10 anos ou mais de idade (R$)": "Income_Median_Per_capita","Unnamed: 5":"With_Access_Med","Unnamed: 6":"Without_Access_Med"})

internet_income_transformed.set_index("Brazilian_States", inplace=True)
internet_income_transformed.head()
```

```python
internet_edu_level = "resources/Percentage_Internet_Usage_Per_Edu_Level.csv"
internet_edu_level_df = pd.read_csv(internet_edu_level)
clean_internet_edu_level_df = internet_edu_level_df.drop([0,1,7,10,14,18,22,23,25,28,30,31,33,36,37,42,43,44])
clean_internet_edu_level_df.head()
```

```python
internet_edu_level_transformed = clean_internet_edu_level_df [["Grandes Regiões, Unidades da Federação e Regiões Metropolitanas", "Total (1)", "Sem instrução e funda-mental incompleto", "Funda-mental completo", "Médio incompleto","Médio completo","Superior incompleto","Superior completo","Não deter-minado"]]
internet_edu_level_transformed= internet_edu_level_transformed.rename(columns={"Grandes Regiões, Unidades da Federação e Regiões Metropolitanas":"Brazilian_States", "Total (1)":"Total_Levels", "Sem instrução e funda-mental incompleto":"Primary_Incomplete", "Funda-mental completo":"Primary_School", "Médio incompleto":"Secondary_Incomplete","Médio completo":"Secondary","Superior incompleto":"Higher_Incomplete","Superior completo":"Higher","Não deter-minado":"undetermined"})

internet_edu_level_transformed.set_index("Brazilian_States", inplace=True)
internet_edu_level_transformed.head()
```

```python
internet_access_income_vs_edu_level = pd.merge(left = internet_income_transformed, right = internet_edu_level_transformed, left_on = "Brazilian_States", right_on = "Brazilian_States")

internet_access_income_vs_edu_level.head()


```
